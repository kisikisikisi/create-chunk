```python
import glob
import os
files = glob.glob("./idata/*")
check = 0
for file in files:
    with open(file ,mode="r", encoding='utf8') as f:
        for i in f.readlines():
            line = []
            for j in i:
                if check == 1:
                    line.append(j)
                if j == '"':
                    check += 1
                if check == 2:
                    check = 0
                    break
            for j in range(len(line)-1):
                print(line[j],end="")
            print("")
```
