```python
#create image data
import glob
import os
files = glob.glob("./idata/*")

with open("img-dm2.lisp",mode="a", encoding='utf8') as f2:
    f2.write("(add-dm-fct '(")

#image0 isa image-data text "image0" type image

for file in files:
    with open(file ,mode="r", encoding='utf8') as f:
        with open("img-dm2.lisp",mode="a", encoding='utf8') as f2:
            s = ""
            tmp = 0
            for line in f.readlines():
                for text in line:
                    if text == '"':
                        tmp += 1
                        if tmp == 1:
                            continue
                    if tmp == 1:
                        s += text
                    elif tmp == 2:
                        break
                f2.write('('+s+' isa image-data text "'+s+'" type image)\n')
                break
        print(file)
        
```

```python
#create label data
import glob
import os
files = glob.glob("./idata/*")
#l-Chrome isa label-data text "Chrome")
#(687-無料 isa inc-label idata 687 lebel l-無料)
for file in files:
    with open(file ,mode="r", encoding='utf8') as f:
        with open("img-dm2.lisp",mode="a", encoding='utf8') as f2:
            ls = []
            for line in f.readlines():
                s = ""
                tmp = 0
                for i in range(len(line)):
                    if line[i] == '"':
                        tmp += 1
                        if tmp == 1:
                            continue
                    if tmp == 2:
                        break
                    elif tmp == 1:
                        s += str(line[i])
                if s == "":
                    break
                ls.append(s)
                f2.write('('+str(ls[0])+"-"+s+" isa inc-label idata "+str(ls[0])+' label l-'+s+')\n')
                f2.write('(l-'+s+' isa label-data text "'+s+'")\n')
        print(file)
        
```

```python
#create time data
import glob
import os
import json
files = glob.glob("./idata/*")
json_open = open('history.json', 'r', encoding='utf8')
json_load = json.load(json_open)

def time_classify(time):
    if time[5] == '4':
        return 'time4'
    elif time[5] == '3':
        return 'time3'
    elif time[5] == '2':
        return 'time2'

for file in files:
    with open(file ,mode="r", encoding='utf8') as f:
        with open("img-dm2.lisp",mode="a", encoding='utf8') as f2:
            s = ""
            tmp = 0
            for line in f.readlines():
                for text in line:
                    if text == '"':
                        tmp += 1
                        if tmp == 1:
                            continue
                    if tmp == 1:
                        s += text
                    elif tmp == 2:
                        break
                for i in range(len(json_load)):
                    if json_load[i]['id'] == s:
                        time = time_classify(json_load[i]['lastVisitTime'])
                        print(time)
                        f2.write('('+s+'-'+time+' isa time-data idata '+s+' time '+time+')\n')
                        break
                break
        print(file)
        
#image0-t0 isa time-data idata image0 time time0
with open("img-dm2.lisp",mode="a", encoding='utf8') as f2:
    f2.write('))')
```
