```python
import glob
import os
files = glob.glob("./idata/*")
check = 0
for file in files:
    with open(file ,mode="r", encoding='utf8') as f:
        for i in f.readlines():
            line = []
            for j in i:
                if check == 1:
                    line.append(j)
                if j == '"':
                    check += 1
                if check == 2:
                    check = 0
                    break
            for j in range(len(line)-1):
                print(line[j],end="")
            print("")
```

```python
#create image data
import glob
import os

files = glob.glob("visicon/idata2/*")
new_file = "declarative memory/img-dm2.lisp"

with open("declarative memory/img-dm2.lisp",mode="w", encoding='utf8') as f2:
    f2.write("(add-dm-fct '(")

#image0 isa image-data text "image0" type image

for i,file in enumerate(files):
    with open(file ,mode="r", encoding='utf8') as f:
        with open(new_file, mode="a", encoding='utf8') as f2:
            s = ""
            tmp = 0
            for line in f.readlines():
                for text in line:
                    if text == '"':
                        tmp += 1
                        if tmp == 1:
                            continue
                    if tmp == 1:
                        s += text
                    elif tmp == 2:
                        break
                f2.write('('+s+' isa image-data text "'+s+'" type image)\n')
                break
        print(file)
        
```

```python
#create label data
import glob
import os

#files = glob.glob("./idata/*")
#l-Chrome isa label-data text "Chrome")
#(687-無料 isa inc-label idata 687 lebel l-無料)
for i,file in enumerate(files):
    with open(file ,mode="r", encoding='utf8') as f:
        with open(new_file, mode="a", encoding='utf8') as f2:
            ls = []
            for line in f.readlines():
                s = ""
                tmp = 0
                for i in range(len(line)):
                    if line[i] == '"':
                        tmp += 1
                        if tmp == 1:
                            continue
                    if tmp == 2:
                        break
                    elif tmp == 1:
                        s += str(line[i])
                if s == "":
                    break
                ls.append(s)
                if ls[0] == s:
                    continue
                f2.write('(l-'+s+' isa label-data text "'+s+'")\n')
                f2.write('('+str(ls[0])+"-"+s+" isa inc-label idata "+str(ls[0])+' label l-'+s+')\n')
        print(file)
        
```

```python
#create time data
import glob
import os
import json
#files = glob.glob("./idata/*")

json_open = open('history.json', 'r', encoding='utf8')
json_load = json.load(json_open)
chunk_ls = []

def time_classify(time):
    t = time_month(time[5])
    create_time_chunk(t)
    return t
    
def time_month(month):
    return 'time'+str(month)

def create_time_chunk(chunk):
    global chunk_ls
    chunk_ls.append(chunk)

for i,file in enumerate(files):
    with open(file ,mode="r", encoding='utf8') as f:
        with open("img-dm.lisp",mode="a", encoding='utf8') as f2:
            s = ""
            tmp = 0
            for line in f.readlines():
                for text in line:
                    if text == '"':
                        tmp += 1
                        if tmp == 1:
                            continue
                    if tmp == 1:
                        s += text
                    elif tmp == 2:
                        break
                for j in range(len(json_load)):
                    if json_load[j]['id'] == s[:3]:
                        time = time_classify(json_load[j]['lastVisitTime'])
                        f2.write('('+s+'-'+time+' isa time-data idata '+s+' time '+time+')\n')
                        break
                break
        print(file)
        
#image0-t0 isa time-data idata image0 time time0
with open("img-dm.lisp",mode="a", encoding='utf8') as f2:
    chunk_ls = list(set(chunk_ls))
    for i in chunk_ls:
        f2.write('('+ i + ' is chunk)\n')
    f2.write('))')
```
